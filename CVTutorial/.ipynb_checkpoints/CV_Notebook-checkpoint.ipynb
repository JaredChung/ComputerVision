# CV2 Tutorial

## Import images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Read image of a tree

```python
img = cv2.imread('Tree.jpg',0)
cv2.imshow('Random Tree',img)
cv2.waitKey(0) # waits for any key
cv2.destroyAllWindows() # destroy window after key
```

Display in matplotlib

```python
fig,ax =plt.subplots(figsize=(7,7))
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.plot([200,300,400],[100,200,300],'c', linewidth=5)
plt.show()
```

```python
cv2.imwrite('ChangedTree.png',img)
```

## Loading Video Source

```python
cap = cv2.VideoCapture(0)
#fourcc = cv2.VideoWriter_fourcc(*'XVID')
#out = cv2.VideoWriter('output.avi',fourcc,20.0,(640,480))
while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow('frame',frame)
    cv2.imshow('frame2',gray)
    # assign key to exit webcam
    if cv2.waitKey(1) & 0xFF == ord('q'): 
        break

cap.release()
cv2.destroyAllWindows()
```

## Drawing on Images

```python
img = cv2.imread('Tree.jpg',1)

cv2.line(img,(0,0),(150,150),(255,255,255),15)
cv2.rectangle(img,(15,25),(200,150),(0,255,0),5)
cv2.circle(img,(400,63), 63, (0,255,0), -1)

pts = np.array([[100,50],[200,300],[700,200],[500,100]], np.int32)
#pts = pts.reshape((-1,1,2))
cv2.polylines(img,[pts], True,(0,255,255),3)

font = cv2.FONT_HERSHEY_SIMPLEX

cv2.putText(img,'Open CV Awesome', (0,130), font,1,(200,244,255))

cv2.imshow('random',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
